```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/My Drive/CIS-5190-Project/Code/cleaned/econ/us_inflation.csv', header = 0)
```

```python
df = df.iloc[:5477]
df
```

```python
df['date'] = pd.to_datetime(df['DATE'])
```

```python
full_date_range = pd.DataFrame({'date': pd.date_range(start='2000-01-01', end='2024-01-01')})
```

```python
full_df = full_date_range.merge(df, on='date', how='left')
```

```python
full_df['T5YIE'] = full_df['T5YIE'].replace('.', np.nan)
```

```python
full_df[['us_inflation']] = full_df[['T5YIE']].fillna(method='ffill')
```

```python
full_df = full_df[['date', 'us_inflation']]
```

```python
full_df.to_csv("daniel_signals.csv")
```
